## Import Lib

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

## Load Coins image change to Binary

```python
src = cv.imread("./conis.png")
gray_img = cv.cvtColor(src,cv.COLOR_BGR2GRAY)
_,bin_img = cv.threshold(gray_img,127,255,cv.THRESH_BINARY)
plt.imshow(bin_img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
```

## My custom

```python
ellipse7x7 = cv.getStructuringElement(cv.MORPH_ELLIPSE,(7,7)) ## filter ขนาด 7x7
eroded_ellipse7x7_img = cv.erode(bin_img, ellipse7x7, iterations = 5) ## ทำซ้ำ 5 รอบ

print("Original Binary")
plt.imshow(bin_img, cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()

print("Result from eroded_ellipse7x7")
plt.imshow(eroded_ellipse7x7_img, cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```
